```python
def factorial(i) -> int:
    if i == 1: return 1
    return i * factorial(i - 1)


factorial(4)
```

```python
from typing import List


def read(n) -> List[int]:
    if n == 0:
        return []

    string = [n]
    string.extend(read(n - 1))
    return string


read(4)
```

```python
from typing import List

result = []


def build(iterable, repeat, previous=None):
    if previous is None:
        previous = []
    if repeat == 0:
        result.append(previous)
        return

    for i in iterable:
        p = previous.copy()
        p.append(i)
        build(iterable, repeat - 1, p)


def product(iterable, repeat) -> List[List[str]]:
    build(iterable, repeat)
    return result


product(['A', 'B', 'C'], 2)
```

```python
def factorial(i) -> int:
    if i == 1: return 1
    return i * factorial(i - 1)


factorial(4)
```

Implementing factorial is easy because it's a pure recursive function.
The result is the return value.

```python
from typing import List


def backtrack(curr: List[str], remaining: List[str], result: List[List[str]]):
    if not remaining:
        result.append(curr.copy())
        return

    for choice in remaining:
        next_remaining = remaining.copy()
        next_remaining.remove(choice)

        curr.append(choice)
        backtrack(curr, next_remaining, result)
        curr.pop()


def get_permutations(iterable: List[str]) -> List[List[str]]:
    result = []
    backtrack([], iterable, result)
    return result


# Example
get_permutations(['A', 'B', 'C', 'D'])
```
